```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn')
sns.set(font_scale=2.5)

%matplotlib inline
```

```python
df_task1 = pd.read_csv('task1.csv')
df_task1['bias'] = df_task1['bias'].map(lambda x: x[2:-2])
df_task1['w1'] = df_task1['weight'].map(lambda x: x[2:-2].split()[0])
df_task1['w2'] = df_task1['weight'].map(lambda x: x[2:-2].split()[1])
df_task1['bias'] = pd.to_numeric(df_task1['bias'])
df_task1['w1'] = pd.to_numeric(df_task1['w1'])
df_task1['w2'] = pd.to_numeric(df_task1['w2'])
df_task1.head()
```

```python
df_task2 = pd.read_csv('task2.csv')
df_task2['bias1'] = df_task2['bias1'].map(lambda x: x[2:-2])
df_task2['bias2'] = df_task2['bias2'].map(lambda x: x[2:-2])
df_task2['w1-1'] = df_task2['weight 1'].map(lambda x: x.split()[0][2:])
df_task2['w1-2'] = df_task2['weight 1'].map(lambda x: x.split()[1][:-2])
df_task2['w2'] = df_task2['weight 2'].map(lambda x: x[2:-2])
df_task2['bias1'] = pd.to_numeric(df_task2['bias1'])
df_task2['bias2'] = pd.to_numeric(df_task2['bias2'])
df_task2['w1-1'] = pd.to_numeric(df_task2['w1-1'])
df_task2['w1-2'] = pd.to_numeric(df_task2['w1-2'])
df_task2['w2'] = pd.to_numeric(df_task2['w2'])
df_task2.head()
```

```python
df_task3 = pd.read_csv('task3.csv')
df_task3['bias1_raw'] = df_task3['bias1'].map(lambda x: x[2:-2]).map(lambda x: x.split('\n'))
df_task3['b1-1'] = df_task3['bias1_raw'].map(lambda x: x[0][:-1])
df_task3['b1-2'] = df_task3['bias1_raw'].map(lambda x: x[1][2:-1])
df_task3['b1-3'] = df_task3['bias1_raw'].map(lambda x: x[2][2:-1])
df_task3['bias2'] = df_task3['bias2'].map(lambda x: x[2:-2])
df_task3['weight1_raw'] = df_task3['weight 1'].map(lambda x: x[2:-2]).map(lambda x: x.split('\n'))
df_task3['w1-1'] = df_task3['weight1_raw'].map(lambda x: x[0].split()[0][:])
df_task3['w1-2'] = df_task3['weight1_raw'].map(lambda x: x[0].split()[1][:-1])
df_task3['w1-3'] = df_task3['weight1_raw'].map(lambda x: x[1][2:-1].split()[0])
df_task3['w1-4'] = df_task3['weight1_raw'].map(lambda x: x[1][2:-1].split()[1])
df_task3['w1-5'] = df_task3['weight1_raw'].map(lambda x: x[2][2:].split()[0])
df_task3['w1-6'] = df_task3['weight1_raw'].map(lambda x: x[2][2:].split()[1])
df_task3['w2-1'] = df_task3['weight 2'].map(lambda x: x[2:-2].split()[0])
df_task3['w2-2'] = df_task3['weight 2'].map(lambda x: x[2:-2].split()[1])
df_task3['w2-3'] = df_task3['weight 2'].map(lambda x: x[2:-2].split()[2])
df_task3['b1-1'] = pd.to_numeric(df_task3['b1-1'])
df_task3['b1-2'] = pd.to_numeric(df_task3['b1-2'])
df_task3['b1-3'] = pd.to_numeric(df_task3['b1-3'])
df_task3['bias2'] = pd.to_numeric(df_task3['bias2'])
df_task3['w1-1'] = pd.to_numeric(df_task3['w1-1'])
df_task3['w1-2'] = pd.to_numeric(df_task3['w1-2'])
df_task3['w1-3'] = pd.to_numeric(df_task3['w1-3'])
df_task3['w1-4'] = pd.to_numeric(df_task3['w1-4'])
df_task3['w1-5'] = pd.to_numeric(df_task3['w1-5'])
df_task3['w1-6'] = pd.to_numeric(df_task3['w1-6'])
df_task3['w2-1'] = pd.to_numeric(df_task3['w2-1'])
df_task3['w2-2'] = pd.to_numeric(df_task3['w2-2'])
df_task3['w2-3'] = pd.to_numeric(df_task3['w2-3'])
df_task3.head()
```
